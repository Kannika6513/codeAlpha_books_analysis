```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://books.toscrape.com/catalogue/page-{}.html"
rating_dict = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

book_data = []

for page in range(1, 51):
    print(f"Scraping page {page}...")
    url = base_url.format(page)
    res = requests.get(url)

    if res.status_code != 200:
        print(f"Could not fetch page {page}")
        continue

    soup = BeautifulSoup(res.text, 'html.parser')
    books = soup.find_all('article', class_='product_pod')

    for book in books:
        title = book.h3.a['title']
        raw_price = book.find('p', class_='price_color').text
        clean_price = raw_price.encode('ascii', 'ignore').decode().replace('Â£', '').strip()
        rating_class = book.p.get('class')[1]
        rating = rating_dict.get(rating_class, 0)

        detail_url = "https://books.toscrape.com/catalogue/" + book.h3.a['href']
        detail_res = requests.get(detail_url)
        detail_soup = BeautifulSoup(detail_res.text, 'html.parser')

        genre = "Unknown"
        breadcrumb = detail_soup.select('ul.breadcrumb li a')
        if len(breadcrumb) > 2:
            genre = breadcrumb[2].text.strip()

            book_data.append({
            "Title": title,
            "Price (GBP)": float(clean_price),
            "Rating": rating,
            "Genre": genre

        })

df = pd.DataFrame(book_data)

df.to_csv("scraped_bookdata.csv", index=False, encoding='utf-8-sig')

print("\n Done! Data saved as scraped_bookdata.csv'")
```
